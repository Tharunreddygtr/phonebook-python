```python
#method  (1)

data = {}

flag = True
while flag:
    try:
        input1 = int(input("choose one option \n(1) Adding contact \n(2) Searching for contact \n(3) Deleting the contact  \n(4) searching by partial name \n(5) close \n "))
        if input1 == 1:
            name = input("Please enter the name of the person: ")
            place = input("Please enter the address of the person : ")
            phone_number = input("please enter the phone number /mobile number: ")
            if input("any alternate number choose \n yes (or) no \n") == "yes":
                alternate_mobile_no = input("Please enter the alternate mobile number")
                data[name] = [name, phone_number, alternate_mobile_no, place]
            else:
                data[name] = [name, phone_number, place]
        elif input1 == 2:
            searching_by_name = input("Please enter the name of the person: ")
            if searching_by_name not in data.keys():
                print("No data found by the provided details...")
            else:
                print("The following data is found by provided details")
                print(data[searching_by_name])
        elif input1 == 3:
            name = input("please enter the name of the person to delete contact details: ")
            del data[name]
        elif input1==4:
            check=True
            searching_partial_name=input("please enter the partial name ")
            for name in data.keys():
                if data[name][0].startswith(searching_partial_name):
                    print("data found by provided partial name")
                    print(data[name])
                    check=False
            if check:
                print("no data found by the provided information")

        elif input1 == 5:
            flag = False

        else:
            print("please enter the valid number from 1 to 4")

    except ValueError:
        print("please enter the integer values between 1 to 4")

    print()

```

```python


# method (3) using two classes

class person:
    def __init__(self,name,phone_number,place):
        self.name=name
        self.phone_number=phone_number
        self.place=place


obj1 = person("tharun", 9381733574, "hyd")
obj2 = person("chintu", 9121427446, "hyd")
obj3 = person("arun", 9381733574, "delhi")
obj4 = person("sai", 9381733574, "hyd")
obj5 = person("yatin", 9389733574, "hyd")
obj6 = person("varun", 9381733575, "kolkata")
obj7 = person("shyam", 9381733589, "hyd")


class phonebook:
    book={}
    @staticmethod
    def adding_contact_to_phonebook(object1):
        phonebook.book[object1.name]=[object1.name,object1.phone_number,object1.place]
        print("provided data is added to the phonebook")


    @classmethod
    def searching_contact(cls,name):
        try:
            found=cls.book[name]
            print("data found by provided name ",found)
        except KeyError:
            print("data not found by the provided name...")

    @classmethod
    def deleting_contact(cls,name):
        if name in cls.book.keys():
            del cls.book[name]
        else:
            print("No data found by the provided details")

    @classmethod
    def available_data(cls):
        print("available data in the phonebook ")
        print(cls.book)

    @classmethod
    def search_by_mobile_no(cls,mobile_no):
        for i in cls.book.keys():
            if cls.book[i][1]==mobile_no:
                print("data found by the provided details")
                print(cls.book[i])
        else:
            print("no data found")



phonebook.adding_contact_to_phonebook(obj1)
phonebook.adding_contact_to_phonebook(obj2)
phonebook.adding_contact_to_phonebook(obj3)
phonebook.adding_contact_to_phonebook(obj4)
phonebook.adding_contact_to_phonebook(obj5)

phonebook.available_data()

phonebook.searching_contact("tharun")

phonebook.search_by_mobile_no(9381733574)


```
